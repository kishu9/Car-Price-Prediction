# Feature Eng

```python
datacar['fueleconomy'] = (0.5 * datacar['citympg']) + (0.5 * datacar['highwaympg'])    #Fuel economy
```

```python
#Binning the Car Companies based on avg prices of each Company.
datacar['price'] = datacar['price'].astype('int')
temptable = datacar.copy()
table = temptable.groupby(['CompanyName'])['price'].mean()
temptable = temptable.merge(table.reset_index(), how='left',on='CompanyName')
bins = [0,10000,20000,40000]
datacar_bin=['Budget','Medium','Highend']
datacar['carsrange'] = pd.cut(temptable['price_y'],bins,right=False,labels=datacar_bin)
datacar.head(27)
```

```python
plt.figure(figsize=(8,6))

plt.title('Price vs Fuel economy')
sns.scatterplot(x = datacar['price'],y = datacar['fueleconomy'],hue=datacar['drivewheel'])
plt.xlabel('Fuel Economy')
plt.ylabel('Price')

plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(35, 5))           #High ranged cars prefer rwd drivewheel with idi or mpfi fuelsystem.

df = pd.DataFrame(datacar.groupby(['fuelsystem','drivewheel','carsrange'])['price'].mean().unstack(fill_value=0))
df.plot.bar()
plt.title('Car Range vs Average Price')
plt.show()
```

# Model Building

```python
Y_train = df_train.pop('price')
X_train = df_train
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
lr = LinearRegression()             #using RFE
lr.fit(X_train,Y_train)

# Subsetting training data for 15 selected columns
rfe = RFE(lr,15)
rfe.fit(X_train, Y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
cols = X_train.columns[rfe.support_]
cols
```

# Model 1:

```python
import statsmodels.api as sm
```

```python
X1 = X_train[cols]
X1_sm = sm.add_constant(X1)

lr_1 = sm.OLS(Y_train,X1_sm).fit()

print(lr_1.summary())
```

```python
vif = pd.DataFrame()       #VIF
vif['Features'] = X1.columns
vif['VIF'] = [variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

Here we get a few variables which have an infinite vif. These variables aren't of use. But manually elimination is time consuming and makes the code unnecessarily long. So let's try and build a model with 10 features this time using RFE.

```python
lr_2 = LinearRegression()

rfe2 = RFE(lr_2,10)
rfe2.fit(X_train, Y_train)
```

```python
list(zip(X_train.columns,rfe2.support_,rfe2.ranking_))
```

```python
supportedcols = X_train.columns[rfe2.support_]
print(supportedcols)
```

# Model 2:

```python
X2 = X_train[supportedcols]
X2_sm = sm.add_constant(X2)

model_2 = sm.OLS(Y_train,X2_sm).fit()

print(model_2.summary())
```

```python
vif = pd.DataFrame()     #VIF
vif['Features'] = X2.columns
vif['VIF'] = [variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

As we see, still there are columns with high VIF. Let us drop column - CompanyName_subaru.

# Model 3:

```python
X3 = X2.drop(['CompanyName_subaru'], axis =1)
X3_sm = sm.add_constant(X3)

model_3 = sm.OLS(Y_train,X3_sm).fit()

print(model_3.summary())
```

```python
vif = pd.DataFrame()       #VIF
vif['Features'] = X3.columns
vif['VIF'] = [variance_inflation_factor(X3.values, i) for i in range(X3.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

# Model 4:

```python
X4 = X3.drop(['enginetype_ohcf'], axis =1)     #drop column -enginetype_ohcf
X4_sm = sm.add_constant(X4)

model_4 = sm.OLS(Y_train, X4_sm).fit()

print(model_4.summary())
```

```python
vif = pd.DataFrame()    #VIF
vif['Features'] = X4.columns
vif['VIF'] = [variance_inflation_factor(X4.values, i) for i in range(X4.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

# Model 5:

```python
X5 = X4.drop(['CompanyName_peugeot'], axis =1)     #dropping CompanyName_peugeot
X5_sm = sm.add_constant(X5)

model_5 = sm.OLS(Y_train,X5_sm).fit()

print(model_5.summary())
```

```python
vif = pd.DataFrame()   #VIF
vif['Features'] = X5.columns
vif['VIF'] = [variance_inflation_factor(X5.values, i) for i in range(X5.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

# Model 6:

```python
X6 = X5.drop(['enginetype_l'], axis =1)      #dropping enginetype_l 
X6_sm = sm.add_constant(X6)

model_6 = sm.OLS(Y_train,X6_sm).fit()

print(model_6.summary())
```

```python
vif = pd.DataFrame()     #VIF
vif['Features'] = X6.columns
vif['VIF'] = [variance_inflation_factor(X6.values, i) for i in range(X6.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

All the VIF values and p-values seem to be in a good range. Also the Adjusted R-squared is 89%. This model is explaining most of the variance without being too complex.
